```python
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as ss
import statsmodels.formula.api as smf
import statsmodels.api as sm
```

```python
data = pd.read_csv('cars.csv')
data.head()
```

```python
data.isna().sum()
```

```python
data['company'] = data['CarName'].apply(lambda x: x.split(' ')[0])
data.head()
```

```python
data.company.nunique()
```

```python
data.drop(['CarName', 'car_ID'], axis=1, inplace=True)
data.head()
```

```python
data['company'].unique()
```

```python
data['company'] = data['company'].replace('maxda', 'mazda')
data['company'] = data['company'].replace('Nissan', 'nissan')
data['company'] = data['company'].replace('porcshce', 'porsche')
data['company'] = data['company'].replace('toyouta', 'toyota')
data['company'] = data['company'].replace(['vokswagen', 'vw'], 'volkswagen')
```

```python
data.company.nunique()
```

```python
data_predictors = data[['price', 'company', 'fueltype', 'aspiration', 'carbody',
                        'drivewheel', 'wheelbase', 'carlength', 'carwidth',
                        'curbweight', 'enginetype', 'cylindernumber',
                        'enginesize', 'boreratio', 'horsepower']]
data_predictors.head()
```

```python
data_predictors.corr()
```

```python
data_predictors.dtypes
```

```python
df_dummy = pd.get_dummies(data=data_predictors[['company', 'fueltype',
                                                'aspiration','carbody',
                                                'drivewheel', 'wheelbase',
                                                'carlength','carwidth',
                                                'curbweight', 'enginetype',
                                                'cylindernumber', 'enginesize',
                                                'boreratio','horsepower',
                                                'price']], drop_first = True)
df_dummy.head()
```

```python
model = smf.ols('price ~ horsepower', data=df_dummy).fit()
model.summary()
```

```python
X = df_dummy.drop(['price'], axis=1)
X = sm.add_constant(X)
y = df_dummy.price
```

```python
model_1 = sm.OLS(y, X).fit()
model_1.summary()
```

```python
cols_to_keep = df_dummy.columns[~df_dummy.columns.str.startswith('company_')]
X = df_dummy[cols_to_keep].drop(['price'], axis=1)
X = sm.add_constant(X)
y = df_dummy.price
```

```python
model_2 = sm.OLS(y, X).fit()
model_2.summary()
```
